A damped sinusoid is a smooth, periodic oscillation with an amplitude that approaches zero as time increases." (wikipedia)

```python
import numpy as np
import matplotlib.pyplot as plt

gamma0 = 2
omega0 = 24
omega1 = 48

def equation(x, omega, gamma):
  return (np.exp((-gamma)*x) * np.cos(omega * x))

x = np.linspace(0, 1, 1000)

y0 = equation(x, omega0, gamma0)
y1 = equation(x, omega1, gamma0)

plt.plot(x, y0) 
plt.plot(x, y1)
```
