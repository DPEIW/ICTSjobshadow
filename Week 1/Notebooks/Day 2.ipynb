```python
#line 16 np.random.rand changed to np.random.randn so that values are picked from within the normal distribution
import numpy as np
import matplotlib.pyplot as plt

gamma0 = 2
omega0 = 24

def equation(x, omega, gamma):
  return (np.exp((-gamma)*x) * np.cos(omega * x))

x = np.linspace(0, 1, 1000)

y0 = equation(x, omega0, gamma0)

x_linear = np.linspace(-1, 1, 1000)
y_comb = y0 + (np.random.randn(1000)*20) 
#change the multiplied value to make it more/less hidden

plt.plot(x_linear, y_comb)
```

```python
from scipy.optimize import least_squares
time_axis = np.linspace(-1, 1, 100)
s = 5*time_axis + 4

plt.plot(time_axis, s)
plt.show
```

```python
n = np.random.randn(100)

d = s + n

plt.scatter(time_axis, s, label ='signal', c = 'orange')
plt.plot(time_axis, d, label = 'data')
plt.legend()
plt.show()
```

```python
def StraightLine(params):
    m = params[0]
    c = params[1]

    return m*time_axis + c

StraightLine([2, 3])
```

```python
def ChiSquared(params):
    t = StraightLine(params)

    chisq = np.sum((d - t)**2)

    return chisq

result = least_squares(ChiSquared, [0, 0])

result
```
